### Created a consolidated, analysis-ready dataset (sales_full), containing transactional, customer, product, store, and currency information in a single table.


```python
sales_data['Order Date']=pd.to_datetime(sales_data['Order Date'])
exchange_rates['Date']=pd.to_datetime(exchange_rates['Date'])
sales_fx =sales_data.merge(
    exchange_rates,
    left_on=['Order Date', 'Currency Code'],
    right_on=['Date', 'Currency'],
    how='left'
)

# Merge master data
sales_full = (
    sales_fx
    .merge(customer_info, on='CustomerKey', how='left')
    .merge(product_info, on='ProductKey', how='left')
    .merge(stores_info, on='StoreKey', how='left')
)
```

### Revenue Standardization Since transactions occurred across multiple currencies, exchange-rate data was integrated to convert revenue into a common USD format. This enables accurate cross-country comparisons and performance benchmarking.

```python
# Create USD-normalised revenue
product_info['Unit Cost USD'] = product_info['Unit Cost USD'].replace('[$,]', '', regex=True)
product_info['Unit Price USD'] = product_info['Unit Price USD'].replace('[$,]', '', regex=True)
sales_full[['Exchange','Quantity','Unit Price USD','Unit Cost USD']] = sales_full[['Exchange','Quantity','Unit Price USD','Unit Cost USD']].apply(
    pd.to_numeric,
    errors='coerce')
sales_full['Revenue_USD'] = (
    sales_full['Quantity'] *
    sales_full['Unit Price USD'] *
    sales_full['Exchange']
)
```

### KPI 1: Store Revenue Efficiency

Metric:

Revenue per Square Meter

Formula:

Revenue_USD / Square_Meters

Purpose:

Evaluate how efficiently each store utilizes physical retail space.


```python
sales_full['revenue_per_Sqft']=(sales_full['Revenue_USD']/sales_full['Square Meters'])

print(sales_full.groupby(['Stores_country', 'Stores_state'])['revenue_per_Sqft'].mean().sort_values(ascending=False))
```

### KPI 2: Recurring growth Month over Month

Metric:

Recurring Revenue Growth monthly

Formula:

((Current Month RevenueâPrevious Month Revenue)/Previous Month Revenue)âÃ100

Purpose:

Evaluate how efficiently store performance for month after month



```python
sales_full['Order Date'] = pd.to_datetime(sales_full['Order Date'])
sales_full['Month'] = sales_full['Order Date'].dt.to_period('M')
monthly_revenue = sales_full.groupby('Month')['Revenue_USD'].sum()
mom_growth = (monthly_revenue.pct_change() * 100)
print(mom_growth)
print(monthly_revenue.rolling(3).mean())
```

### Observation:

As from jan to feb the revenue increased by 37.09 percent approx, but then revenue dropped by 62.39 percent for the month march. Now the reason could be change in demand of the product and since its a new store opening, it will take some time to gain traction, as its very normal for any store business, as you can see for the month of MAY revenue increased by 407.49 percent, but yes sharp drops are there. ## KPI 3: Store Age Performance

Stores were segmented into:

- 0â5 Years
- 5â10 Years
- 10â15 Years
- 15â20 Years
- 20+ Years

The objective was to evaluate whether mature stores outperform newer locations.

```python
sales_full['Open Date'] = pd.to_datetime(sales_full['Open Date'])

sales_full['Store_Age'] = (
    pd.Timestamp.today().year -
    sales_full['Open Date'].dt.year
)
sales_full['Store_Age_Bucket'] = pd.cut(
    sales_full['Store_Age'],
    bins=[0, 5, 10, 15, 20, 100],
    labels=[
        '0â5 years',
        '5â10 years',
        '10â15 years',
        '15â20 years',
        '20+ years'
    ],
    right=False
)


print(
    sales_full[['Open Date', 'Store_Age', 'Store_Age_Bucket']].head()
)
```

```python
print(
    sales_full.groupby('Store_Age_Bucket')['Revenue_USD']
    .sum()
    .sort_values(ascending=False).round(2)
)
```

```python
print(
    sales_full.groupby('Store_Age_Bucket')['Revenue_USD']
    .mean()
    .round(2)
    .sort_values(ascending=False))
```

### Observation:
Older stores have higher average revenue, whereas Mature stores perform better, Newer stores have high average revenue	hence the stores Expansion strategy is effective. Mid-age stores are underperforming, hence Operational optimization may be needed.
